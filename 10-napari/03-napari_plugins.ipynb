# napari: plugins

Just like in other software like Fiji, napari's capabilities are enhanced by an eco-system of plugins developed by researchers around the world.

````python
import napari
from napari.utils import nbscreenshot
````

## The napari hub

You can find officially released plugins on the [napari-hub](https://www.napari-hub.org/). The site can be searched by key-words and you can find there extensive information about how to use them, where to find more information etc.

Once you have found the plugin of your dreams you can come back to napari and use ```Plugins -> Install/Uninstall Pugins...``` to access the plugin interface:

<img src='../illustrations/plugin_interface.png'>

In this window you can search for your plugin in the ```filter...``` field. Once you have found the right one, you can just click on the install button. Note that because of the fast-evolving nature of napari, it's not 100% guaranteed that the installation will work. In fact, clicking on install, just runs a ```pip install a_package``` command. So if something doesn't work properly, just open a command line and try to install manually.

## Plugins not on the hub

It is not a requirement to publish a plugin on the napari-hub. In particular when a plugin has been developed for a specific project, or a plugin is in its early days, it's probable that the plugin will only be available on PyPi or on GitHub. In that case you need to install it with a regular ```pip``` command.


## An example

Let's install the [napari-animation](https://pypi.org/project/napari-animation/) plugin. We can use the interface as described above or just run a pip command:

````python
pip install napari-animation
````

Now when we check out the plugin menu we see:
<img src='../illustrations/plugin_animation.png'>

We have now an ```animation: Wizard``` option. If we click on this, it will add to the napari window the interface of the plugin, so that we can interact with our data. In the case of the animation plugin this looks like this:

````python
viewer = napari.Viewer()
````

````python
nbscreenshot(viewer=viewer)
````
